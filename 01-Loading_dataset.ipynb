```python
host = "localhost"
user = "root"
password = ""
database = "project_dataset"

# Establish Connection

conn = pymysql.connect(host=host, user=user, password=password, database=database)
```

```python
# sql query

query = "SELECT id, name FROM brands ORDER BY name"
df = pd.read_sql(query, conn)
df.to_csv("brands.csv")
```

```python
query = "SELECT PW.id, product, type, PW.sizeinmm AS name, P.billingin, PW.weight FROM products AS P, productweights AS PW "
query += "WHERE P.id = PW.pid ORDER BY P.product, PW.sizeinmm"
df = pd.read_sql(query, conn)
df.to_csv("products.csv")
```

```python
query = "SELECT month, year, brandid, pwid, brandname, productname, name, SUM(weight) AS weight FROM(SELECT MONTH(O.createdon) AS month, YEAR(O.createdon) AS year, B.id AS brandid, PW.id AS  pwid, B.name AS brandname, P.product AS productname, PW.sizeinmm AS name, OD.weight FROM orders AS O, orderdetails AS OD, products AS P, productweights AS PW, brands AS B WHERE O.id = OD.oid AND OD.pid = P.id AND PW.id = OD.pwid AND B.id = OD.brandid) AS SRC GROUP BY month, year, brandid, pwid, brandname, name ORDER BY year, month"
df = pd.read_sql(query, conn)
df.to_csv("products.csv")
```
